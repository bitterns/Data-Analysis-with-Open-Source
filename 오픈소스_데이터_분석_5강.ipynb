## 5-2 JSON 형식 저장

```python
### 라이브러리 임포트
import json
import pandas as pd

data = {
    "이름": "홍길동",
    "나이": 25,
    "거주지": "서울",
    "관심사": ["프로그래밍", "데이터 분석", "여행"]
}

## json.dump를 이용한 저장
# indent: 들여쓰기 크기
# ensure_ascii: 디폴트 설정인 아스키코드 저장 방식 비활성화
with open('output.json', mode='w', encoding='utf-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

## DataFrame을 이용한 저장
# orient: string 형식 결정
#    records 는 딕셔너리를 요소로 갖는 리스트. 디폴트 값은 columns
# force_ascii: ensure_ascii 와 동일

# json.dump 와 달리 관심사 리스트가 3개의 행으로 분리되었음에 주목
df = pd.DataFrame(data)
df.to_json('output_df.json', orient='records', indent=4, force_ascii=False)

# 데이터를 리스트 형태로 넣고, orient 를 columns 로 지정했을 때 데이터 형태에 주목
df_list = pd.DataFrame([data])
df_list.to_json('output_df_col.json', orient='columns', indent=4, force_ascii=False)
```
